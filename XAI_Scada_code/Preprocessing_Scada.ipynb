```python
import pandas as pd
dataframe = pd.read_csv('dataset_processed.csv', header=0)
dataframe
```

```python
dataframe = dataframe.dropna()
```

```python
labels = dataframe.iloc[:, 10]
labels = labels.where(labels=='No', 1)
labels = labels.where(labels==1, 0)

data = dataframe.iloc[:, 0:10]
```

```python
Data_frame = pd.concat([data , labels], axis = 1)
```

```python
#Chia theo ty le 8:2
Train = pd.concat([Data_frame.iloc[0:1600,:],Data_frame.iloc[2000:12604,:]])
Test = pd.concat([Data_frame.iloc[1600:2000,:],Data_frame.iloc[12604:,:]])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(Train)

#Chuẩn hóa dữ liệu
Train = scaler.transform(Train)
Test = scaler.transform(Test)
```

```python
Train = pd.DataFrame(Train)
Test = pd.DataFrame(Test)
```

```python
A = dataframe.iloc[:, 0:11]
Train.columns = A.columns
Test.columns = A.columns
```

```python
Train.to_excel("traindata.xlsx",index = 0)
Test.to_excel("testdata.xlsx",index = 0)
```
